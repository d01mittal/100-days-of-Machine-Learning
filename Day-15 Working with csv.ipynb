# NOTE- If we don't have the columns name in the data then sometimes the first row becomes the columns name then to overcome this, 'read_csv' function has a parameter 'names' in which we can pass a list of column names

# NOTE- Sometimes we don't want the indexes in the data and we want to make  a specific column as our index then we can use :-

```python
pd.read_csv('aug_train.csv',index_col='enrollee_id')
```

# NOTE- Sometimes, our first row of the data is the column name but our 'read_csv' function makes those column names as first row so in those case, we will do like this:-

```python
pd.read_csv('aug_train.csv',header=1)
```

# Sometimes, we don't want to read all the columns of the data i.e., we want to get only the specific columns then we will pass a list to the parameter 'usecols' of the column names which we want:-

```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','city','gender'])
```

# Skip rows

```python
pd.read_csv('aug_train.csv',skiprows=[0,5])
```

# Getting only n rows

```python
pd.read_csv('aug_train.csv',nrows=100)
```

# Sometimes in some rows there are some column entries missing so not to throw an error for this we write like this:-

**pd.read_csv('name-of-file',error_bad_lines=False)**

# Changing the dtypes of a column

```python
pd.read_csv('aug_train.csv',dtype={'target':int})
```

# To get date in not string format we will do like this:-

**pd.read_csv('name-of-file',parse_dates=['name-of-column'])**
