```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('mnist.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
sns.heatmap(X.iloc[5].values.reshape(28,28))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(X,y)
```

```python
rf.feature_importances_
```
